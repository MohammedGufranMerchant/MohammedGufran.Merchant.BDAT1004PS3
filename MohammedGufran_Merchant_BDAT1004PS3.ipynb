Occupations

Step 1. Import the necessary libraries

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called users

Step 4. Discover what is the mean age per occupation

Step 5. Discover the Male ratio per occupation and sort it from the most to the least

Step 6. For each occupation, calculate the minimum and maximum ages

Step 7. For each combination of occupation and sex, calculate the mean age

Step 8. For each occupation present the percentage of women and men

```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 3: Assign it to a variable called users

# Step 4: Discover the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
def male_ratio(gender_series):
    return (gender_series == 'M').sum() / gender_series.count()

male_ratio_per_occupation = users.groupby('occupation')['gender'].agg(male_ratio)
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: For each occupation, calculate the minimum and maximum ages
age_range_per_occupation = users.groupby('occupation')['age'].agg([min, max])

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: For each occupation, present the percentage of women and men
def gender_percentage(gender_series):
    return (gender_series == 'F').sum() / gender_series.count() * 100

gender_percentage_per_occupation = users.groupby('occupation')['gender'].agg(gender_percentage)

# Print the results
print("Mean age per occupation:")
print(mean_age_per_occupation)

print("\nMale ratio per occupation (sorted):")
print(sorted_male_ratio)

print("\nAge range per occupation:")
print(age_range_per_occupation)

print("\nMean age per occupation and gender:")
print(mean_age_per_combination)

print("\nPercentage of women and men per occupation:")
print(gender_percentage_per_occupation)
```

```python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Import the dataset
data = pd.read_csv('wind.txt', delim_whitespace=True, parse_dates=[[0, 1, 2]])

# Step 3: Replace the first 3 columns with a proper datetime index
data = data.rename(columns={'Yr_Mo_Dy': 'Date'})
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Step 4: Function to fix the year 2061 and apply it
def fix_year(year):
    if year >= 2000:
        return year - 100
    else:
        return year

data['Yr'] = data.index.year
data['Yr'] = data['Yr'].apply(fix_year)
data.index = data.index.to_period('D')

# Step 6: Compute the number of missing values for each location
missing_values = data.isnull().sum()

# Step 7: Compute the number of non-missing values in total
total_non_missing = data.notnull().sum().sum()

# Step 8: Calculate the mean windspeeds over all locations and times
mean_windspeed = data.mean().mean()

# Step 9: Calculate min, max, mean, and standard deviation of windspeeds for each location
loc_stats = data.agg(['min', 'max', 'mean', 'std'], axis=0)

# Step 10: Calculate min, max, mean, and standard deviation of windspeeds for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])

# Step 11: Calculate the average windspeed in January for each location
#january_avg = data[data.index.month == 1].groupby(data.index.year).mean()

# Step 12: Downsample to yearly frequency for each location
yearly_data = data.resample('Y').mean()

# Step 13: Downsample to monthly frequency for each location
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency for each location
weekly_data = data.resample('W-Mon').mean()

# Step 15: Calculate min, max, mean, and standard deviation of windspeeds for each week for the first 52 weeks
weekly_stats = data[data.index.weekofyear <= 52].resample('W-Mon').agg(['min', 'max', 'mean', 'std'])

# Print the results of each step as needed
print("Step 6:")
print(missing_values)
print("\nStep 7:")
print(total_non_missing)
print("\nStep 8:")
print(mean_windspeed)
print("\nStep 9:")
print(loc_stats)
print("\nStep 10:")
print(day_stats)
print("\nStep 11:")
#print(january_avg)
print("\nStep 12:")
print(yearly_data)
print("\nStep 13:")
print(monthly_data)
print("\nStep 14:")
print(weekly_data)
print("\nStep 15:")
print(weekly_stats)
```

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
first_10_entries = chipo.head(10)
print("First 10 entries:\n", first_10_entries)

# Step 5: Number of observations in the dataset
num_observations = chipo.shape[0]
print("Number of observations:", num_observations)

# Step 6: Number of columns in the dataset
num_columns = chipo.shape[1]
print("Number of columns:", num_columns)

# Step 7: Print the name of all the columns
column_names = chipo.columns
print("Column names:", column_names)

# Step 8: How is the dataset indexed?
index_type = chipo.index
print("Index type:", index_type)

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("Most-ordered item:", most_ordered_item)

# Step 10: Number of items ordered for the most-ordered item
most_ordered_item_quantity = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("Number of items ordered:", most_ordered_item_quantity)

# Step 11: Most ordered item in the choice_description column
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()
print("Most-ordered choice description:", most_ordered_choice_description)

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()
print("Total items ordered:", total_items_ordered)

# Step 13: Convert item price to float, check type, apply lambda, check type
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
item_price_type_before = chipo['item_price'].dtype
chipo['item_price'] = chipo['item_price'].apply(lambda x: round(x, 2))
item_price_type_after = chipo['item_price'].dtype
print("Item price type before:", item_price_type_before)
print("Item price type after:", item_price_type_after)

# Step 14: Total revenue
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Total revenue:", revenue)

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()
print("Number of orders:", num_orders)

# Step 16: Average revenue per order
average_revenue_per_order = revenue / num_orders
print("Average revenue per order:", average_revenue_per_order)

# Step 17: Number of different items sold
num_different_items_sold = chipo['item_name'].nunique()
print("Number of different items sold:", num_different_items_sold)
```

##Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Import the dataset from the provided file
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Step 3: Create the line plot
plt.figure(figsize=(10, 6))

# Plot marriages per capita
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per capita', marker='o')

# Plot divorces per capita
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per capita', marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

##Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Import the dataset from the provided file
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Step 3: Filter data for the years 1900, 1950, and 2000
years_of_interest = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years_of_interest)]

# Step 4: Create the vertical bar chart
plt.figure(figsize=(8, 6))

# Plot marriages per capita
plt.bar(filtered_data['Year'] - 0.2, filtered_data['Marriages_per_1000'], width=0.4, label='Marriages per capita')

# Plot divorces per capita
plt.bar(filtered_data['Year'] + 0.2, filtered_data['Divorces_per_1000'], width=0.4, label='Divorces per capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks(years_of_interest)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

##Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Import the dataset from the provided file
data = pd.read_csv('actor_kill_counts.csv')

# Step 3: Sort the actors by their kill count in descending order
sorted_data = data.sort_values(by='Count', ascending=False)

# Step 4: Create the horizontal bar chart
plt.figure(figsize=(10, 8))

plt.barh(sorted_data['Actor'], sorted_data['Count'])

# Add labels and title
plt.xlabel('Number of Kills')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.yticks(sorted_data['Actor'])

# Show the actor names on the bars
for index, value in enumerate(sorted_data['Count']):
    plt.text(value, index, str(value))

# Show the plot
plt.tight_layout()
plt.show()
```

##Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Import the dataset from the provided file
data = pd.read_csv('roman-emperor-reigns.csv')

# Step 3: Calculate the number of emperors assassinated and the number of emperors not assassinated
assassinated_count = data[data['Cause_of_Death'] == 'Assassinated']['Emperor'].count()
not_assassinated_count = data[data['Cause_of_Death'] != 'Assassinated']['Emperor'].count()

# Step 4: Create the pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassinated_count, not_assassinated_count]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title
plt.title('Fraction of Roman Emperors Assassinated')

# Show the pie chart
plt.show()
```
